```python
import json
```

```python
with open('mutualFundsList.json', 'r') as f:
    mutual_funds = json.loads(f.read())
```

```python
codes, names = [], []
```

```python
for x in range(len(mutual_funds)):
    codes.append(mutual_funds[x]['schemeCode'])
    names.append(mutual_funds[x]['schemeName'])
```

```python
len(codes), len(names)
```

```python
len(set(codes)), len(set(names))
```
